### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch    
from sklearn.cluster import AgglomerativeClustering
```

```python
# Normalization Function

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    print(x)
    return x
```

```python
crime_norm = norm_func(crime)
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='complete'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='average'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='single'))
```

```python
# create clusters

hc = AgglomerativeClustering(n_clusters=3, metric = 'euclidean', linkage = 'complete',distance_threshold=None)
```

```python
y_hc = hc.fit_predict(crime_norm)
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
crime.drop(['h_clusterid'],axis=1,inplace=True)
crime.head()
```

```python
# Standardisation function 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = pd.DataFrame(scaler.fit_transform(crime),columns = ['Murder','Assault','UrbanPop','Rape'] )
scaled_crime_df
```

```python
scaled_crime_df.info()
```

```python
# Elbow Method

wcss = []
for i in range(1, 10):   
    kmeans = KMeans(n_clusters=i,random_state=42 )
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)
    
print(wcss)    
plt.plot(range(1, 10), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm for 3 & 4 Clusters

# 3 Clusters

from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
crime['clusterid_new'] = clusters_new.labels_
crime
```

```python
crime.sort_values(by = "clusterid_new")
```

```python
#these are standardized values

clusters_new.cluster_centers_
```

```python
crime.groupby('clusterid_new').mean()
```

```python
# Quality of clusters is expressed in terms of Silhoutte score

from sklearn.metrics import silhouette_score
score =silhouette_score(scaled_crime_df, clusters_new.labels_ , metric='euclidean')
score
```

```python
# Let's check for 4 clusters based on Elbow Method

from sklearn.cluster import KMeans
clusters_new1 = KMeans(4, random_state=42)
clusters_new1.fit(scaled_crime_df)
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
crimedb = pd.read_csv("crime_data.csv")
```

```python
crimedb = crimedb.drop('Unnamed: 0', axis=1)
crimedb.head()
```

```python
crime_val = crimedb.values
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(crime_val)
X
```

```python
from sklearn.neighbors import NearestNeighbors

nearest_neighbors = NearestNeighbors(n_neighbors=11)  # Atleast 5 is better
neighbors = nearest_neighbors.fit(X)

distances, indices = neighbors.kneighbors(X)  # For every point, what is neighbour distance

print(distances)
distances = np.sort(distances[:,10], axis=0)

fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")
plt.grid()
```

```python
from kneed import KneeLocator

i = np.arange(len(distances))
knee = KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')

fig = plt.figure(figsize=(5, 5))
knee.plot_knee()
plt.xlabel("Points")
plt.ylabel("Distance")
plt.grid()

print(distances[knee.knee])
```

```python
dbscan = DBSCAN(eps=1.7, min_samples=4)  
clusters = dbscan.fit_predict(X)
set(clusters)
```

```python
# We got only 2 clusters with one of them outliers with eps 1.7, so trying different eps value for more clusters

dbscan = DBSCAN(eps=1.2, min_samples=4)  
clusters = dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
set(clusters)
```

```python
crimedb['cluster']=dbscan.labels_
crimedb.head()
```

```python
crimedb.groupby('cluster').mean()
```

```python
from sklearn.metrics import silhouette_score
score =silhouette_score(X, dbscan.labels_ , metric='euclidean')
```
